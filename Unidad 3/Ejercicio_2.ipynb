Problema 2: Distribución Exponencial

```python
import numpy as np
import matplotlib.pyplot as plt


lmbda = 1 / 15
total_time = 12 * 60
interarrival_times = []
current_time = 0

while current_time < total_time:
    T = -np.log(np.random.uniform()) / lmbda
    current_time += T
    if current_time < total_time:
        interarrival_times.append(T)
plt.hist(interarrival_times, bins=20, density=True, alpha=0.6, color='pink')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Densidad')
plt.title('Distribución de tiempos entre llegadas de vehículos')
plt.grid()
plt.show()
print("Primeros 10 tiempos entre llegadas:", interarrival_times[:10])
```

##Conclusión del código
El código simula los tiempos entre llegadas de vehículos durante una jornada de 12 horas, asumiendo que estos ocurren de manera aleatoria pero siguiendo un patrón típico de eventos independientes.

###Generación de datos:
Se calculaSe calculan los tiempos entre llegadas utilizando valores aleatorios y se suman progresivamente hasta completar el tiempo total.

###Visualización:
Se crea un histograma que muestraSe crea un histograma que muestra cómo se distribuyen los tiempos entre llegadas, donde la mayoría de los intervalos son cortos, pero en ocasiones hay tiempos más largos entre llegadas.

###Interpretación:
El gráfico representa un patrón común en sistemas donde los eventos ocurren de manera aleatoria pero con una frecuencia promedio. Este tipo de simulación es útil para analizar el tráfico, la logística y la gestión de filas en distintos escenarios.
