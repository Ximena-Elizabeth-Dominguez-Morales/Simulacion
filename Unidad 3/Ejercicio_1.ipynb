Problema 1: Distribución Binomial

```python

import numpy as np
import matplotlib.pyplot as plt

n_pedidos = 20
p_exito = 0.7
semana = 7

pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, semana)
print("Pedidos satisfechos cada día de la semana:", pedidos_satisfechos)

plt.figure(figsize=(8, 5))
plt.bar(range(1, semana+1), pedidos_satisfechos, color='violet', alpha=0.7)
plt.xlabel("Día")
plt.ylabel("Pedidos satisfechos")
plt.title("Simulación de Pedidos Satisfechos en una Semana")
plt.show()
```

##Conclusión
Este código simula la cantidad de pedidos satisfechos en una tienda a lo largo de una semana, asumiendo que cada día se reciben 20 pedidos y que cada uno tiene un 70% de probabilidad de éxito.

###Generación de datos:
Se usa una distribución binomial para modelar cuSe utiliza una distribución binomial para modelar cuántos pedidos se completan con éxito cada día.

###Visualización:
Se crea un gráfico de barras donde cada barra representa la cantidad de pedidos satisfechos en un día de la semana.

###Interpretación:
El resultado muestra que , aunque laEl resultado muestra que, aunque la mayoría de los días se logran completar un número alto de pedidos, hay variabilidad debido a la naturaleza aleatoria del proceso.

Este tipo de simulación es útil para evaluar la eficiencia de un sistema de entregas y entender cómo varía el cumplimiento de los pedidos a lo largo del tiempo.
