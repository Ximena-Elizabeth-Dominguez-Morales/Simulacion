Problema 2: Distribución Exponencial

```python
import numpy as np
import matplotlib.pyplot as plt


lmbda = 1 / 15
total_time = 12 * 60
interarrival_times = []
current_time = 0

while current_time < total_time:
    T = -np.log(np.random.uniform()) / lmbda
    current_time += T
    if current_time < total_time:
        interarrival_times.append(T)
plt.hist(interarrival_times, bins=20, density=True, alpha=0.6, color='pink')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Densidad')
plt.title('Distribución de tiempos entre llegadas de vehículos')
plt.grid()
plt.show()
print("Primeros 10 tiempos entre llegadas:", interarrival_times[:10])
```
