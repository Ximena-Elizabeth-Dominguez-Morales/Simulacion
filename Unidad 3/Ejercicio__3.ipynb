Problema 3: Distribución LogNormal

```python
import numpy as np
import matplotlib.pyplot as plt


mu = 10
sigma = 2
num_clients = 50
time_waits = np.random.lognormal(mean=np.log(mu), sigma=sigma, size=num_clients)
plt.hist(time_waits, bins=20, density=True, alpha=0.6, color='turquoise')
plt.xlabel('Tiempo de espera (minutos)')
plt.ylabel('Densidad')
plt.title('Distribución de tiempos de espera de clientes')
plt.grid()
plt.show()
print("Primeros 10 tiempos de espera:", time_waits[:10])
```
